```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/noahgift/regression-concepts/master/height-weight-25k.csv")
df.head()
```

```python
import seaborn as sns
import numpy as np
```

```python
sns.lmplot("Height-Inches", "Weight-Pounds", data=df)
```

```python
sns.jointplot(x="Height-Inches", y="Weight-Pounds", data=df)
```

```python
from sklearn.model_selection import train_test_split
y = df["Weight-Pounds"].values
print(y.shape)
y = y.reshape(-1, 1)
print(y.shape)
```

```python
X = df["Height-Inches"].values
X = X.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(X_train, y_train)
y_predicted = lm.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_predicted, y_test))

rms
```

```python
# process in PyTorch

x_train = np.array(X_train, dtype=np.float32)
x_train = x_train.reshape(-1, 1)
y_train = np.array(y_train, dtype=np.float32)
y_train = y_train.reshape(-1, 1)

x_test = np.array(X_test, dtype=np.float32)
x_test = x_test.reshape(-1, 1)
y_test = np.array(y_test, dtype=np.float32)
y_test = y_test.reshape(-1, 1)
```
